```python
import numpy as np

# Таблиця значень функції
x_values = np.array([2.4, 2.6, 2.8, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2, 4.4, 4.6])
y_values = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

# Крок таблиці
h = x_values[1] - x_values[0]

# Обчислення кінцевих різниць
def finite_differences(y):
    n = len(y)
    delta_y = np.zeros((n, n))
    delta_y[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            delta_y[i, j] = delta_y[i + 1, j - 1] - delta_y[i, j - 1]

    return delta_y

# Отримуємо таблицю кінцевих різниць
delta_table = finite_differences(y_values)

# Виводимо таблицю кінцевих різниць для наочності
print("Таблиця кінцевих різниць:")
np.set_printoptions(precision=4, suppress=True)
print(delta_table)

# Значення кінцевих різниць для першої та другої похідних
#Змінити індекс відповідно рядка, який використовуєте (нумерація починається з нуля)
delta_y1 = delta_table[7, 1]  # Δy1
delta2_y1 = delta_table[7, 2]  # Δ²y1
delta3_y1 = delta_table[7, 3]  # Δ³y1
delta4_y1 = delta_table[7, 4]  # Δ⁴y1

# Функція для обчислення першої похідної за формулою Ньютона
def first_derivative(delta_y1, delta2_y1, delta3_y1, delta4_y1, h):
    return (delta_y1 - delta2_y1 / 2 + delta3_y1 / 3 - delta4_y1 / 4) / h

# Функція для обчислення другої похідної за формулою Ньютона
def second_derivative(delta2_y1, delta3_y1, delta4_y1, h):
    return (delta2_y1 - delta3_y1  + 11*delta4_y1 / 12) / h**2

# Обчислюємо першу та другу похідні в точці x = 3.8
y8_prime = first_derivative(delta_y1, delta2_y1, delta3_y1, delta4_y1, h)
y8_double_prime = second_derivative(delta2_y1, delta3_y1, delta4_y1, h)

# Округляємо до точності 0.001
y8_prime = round(y8_prime, 3)
y8_double_prime = round(y8_double_prime, 3)

# Виводимо результати
print(f"\nПерша похідна y'(3.8) ≈ {y8_prime}")
print(f"Друга похідна y''(3.8) ≈ {y8_double_prime}")
```

```python
import numpy as np

x_values = np.array([2.4, 2.6, 2.8, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2, 4.4, 4.6])
y_values = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

h = x_values[1] - x_values[0]
i = np.where(x_values == 3.8)[0][0]  # індекс точки 3.8

# Перша похідна (центральна різниця)
first_derivative = (y_values[i+1] - y_values[i-1]) / (2 * h)

# Друга похідна (центральна різниця)
second_derivative = (y_values[i+1] - 2*y_values[i] + y_values[i-1]) / (h**2)

print(f"Перша похідна при x = 3.8: {first_derivative:.4f}")
print(f"Друга похідна при x = 3.8: {second_derivative:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Дані
x_values = np.array([2.4, 2.6, 2.8, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2, 4.4, 4.6])
y_values = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

# Крок
h = x_values[1] - x_values[0]

# Похідні за центральною різницею (всередині масиву)
first_derivative = np.zeros_like(y_values)
second_derivative = np.zeros_like(y_values)

for i in range(1, len(x_values)-1):
    first_derivative[i] = (y_values[i+1] - y_values[i-1]) / (2*h)
    second_derivative[i] = (y_values[i+1] - 2*y_values[i] + y_values[i-1]) / (h**2)

# Інтерполяція для побудови гладких ліній
x_fine = np.linspace(x_values[0], x_values[-1], 500)
y_fine = np.interp(x_fine, x_values, y_values)
first_fine = np.interp(x_fine, x_values, first_derivative)
second_fine = np.interp(x_fine, x_values, second_derivative)

# Побудова графіків
plt.figure(figsize=(12, 8))

# Функція
plt.subplot(3, 1, 1)
plt.plot(x_fine, y_fine, color='blue', label='Функція y(x)')
plt.scatter(x_values, y_values, color='red')
plt.title('Функція y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Перша похідна
plt.subplot(3, 1, 2)
plt.plot(x_fine, first_fine, color='orange', label="Перша похідна y'(x)")
plt.title("Перша похідна y'(x)")
plt.xlabel('x')
plt.ylabel("y'")
plt.legend()
plt.grid(True)

# Друга похідна
plt.subplot(3, 1, 3)
plt.plot(x_fine, second_fine, color='green', label="Друга похідна y''(x)")
plt.title("Друга похідна y''(x)")
plt.xlabel('x')
plt.ylabel("y''")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```
